```python
"""
Improved Math Model Evaluation - Fixed Answer Extraction
"""

# ============================================================
# STEP 1: Install Required Packages
# ============================================================

!pip install -q datasets tqdm pandas

# ============================================================
# STEP 2: Import Libraries
# ============================================================

import torch
import json
import re
import os
from tqdm import tqdm
from datetime import datetime
import pandas as pd

# ============================================================
# STEP 3: Improved Answer Extraction Functions
# ============================================================

def extract_answer_from_gsm8k_format(text):
    """Extract answer from GSM8K #### format"""
    if "####" in text:
        answer = text.split("####")[-1].strip()
        # Clean up the answer
        answer = answer.replace(",", "").strip()
        return answer
    return None

def extract_boxed_answer(text):
    """Extract answer from \\boxed{} format (MATH dataset)"""
    # Try to find ALL boxed answers and take the LAST one (final answer)
    # Pattern 1: Handle nested braces
    pattern = r'\\boxed\{([^{}]*(?:\{[^{}]*\}[^{}]*)*)\}'
    matches = re.findall(pattern, text)
    if matches:
        # Take the last boxed answer (final answer)
        final = matches[-1].strip()
        # Clean any remaining LaTeX
        final = final.replace('\\textbf{', '').replace('}', '').replace('\\text{', '')
        return final

    # Fallback to simpler pattern
    pattern = r'\\boxed\{([^}]+)\}'
    matches = re.findall(pattern, text)
    if matches:
        final = matches[-1].strip()
        final = final.replace('\\textbf{', '').replace('}', '').replace('\\text{', '')
        return final

    return None

def extract_answer_from_response(text, dataset_type="gsm8k"):
    """
    Extract answer from model response based on dataset type
    Priority order for GSM8K:
    1. #### format
    2. \\boxed{} format
    3. "The answer is X" patterns
    4. Last number in text (least reliable)
    """
    text = str(text)

    # First try #### format (GSM8K style)
    gsm8k_answer = extract_answer_from_gsm8k_format(text)
    if gsm8k_answer:
        return gsm8k_answer

    # Try boxed format (MATH dataset style) - prioritize this for math models
    boxed = extract_boxed_answer(text)
    if boxed:
        # Clean LaTeX formatting from boxed answer
        boxed_clean = boxed.replace('\\textbf{', '').replace('}', '').replace('\\text{', '')
        # Extract number from cleaned boxed answer
        num_match = re.search(r'([+-]?[\d,]+\.?\d*)', boxed_clean)
        if num_match:
            return num_match.group(1).replace(",", "").strip()
        return boxed_clean.strip()

    # Try common answer patterns with more flexibility
    patterns = [
        r'[Tt]he answer is:?\s*\$?\\?textbf\{([^}]+)\}',  # LaTeX bold
        r'[Tt]he answer is:?\s*\$?([+-]?[\d,]+\.?\d*)\$?',
        r'[Aa]nswer:?\s*\$?([+-]?[\d,]+\.?\d*)\$?',
        r'=\s*\$?([+-]?[\d,]+\.?\d*)\$?\s*(?:\.|$)',
        r'[Tt]herefore,?\s*(?:the answer is)?\s*\$?([+-]?[\d,]+\.?\d*)\$?',
        r'[Ff]inal answer:?\s*\$?([+-]?[\d,]+\.?\d*)\$?',
    ]

    for pattern in patterns:
        match = re.search(pattern, text)
        if match:
            answer = match.group(1).replace(",", "").strip()
            # Clean any remaining LaTeX
            answer = answer.replace('\\textbf{', '').replace('}', '')
            return answer

    # Last resort: find the last number mentioned
    # Remove common non-answer numbers (like step numbers, choice labels)
    cleaned_text = re.sub(r'[Ss]tep \d+', '', text)
    cleaned_text = re.sub(r'\([A-E]\)', '', cleaned_text)  # Remove (A), (B), etc.
    numbers = re.findall(r'([+-]?[\d,]+\.?\d*)', cleaned_text)
    if numbers:
        # Return last number, cleaned
        return numbers[-1].replace(",", "").strip()

    return None

def normalize_numerical_answer(answer):
    """
    Normalize numerical answer for comparison
    Handles: integers, floats, fractions, percentages
    """
    if answer is None:
        return None

    answer = str(answer).strip()

    # Remove common formatting
    answer = answer.replace('$', '').replace(',', '').replace(' ', '')

    # Handle percentages
    if '%' in answer:
        try:
            return float(answer.replace('%', '')) / 100
        except:
            pass

    # Handle fractions
    if '/' in answer and answer.count('/') == 1:
        try:
            num, denom = answer.split('/')
            return float(num) / float(denom)
        except:
            pass

    # Try to convert to float
    try:
        return float(answer)
    except:
        # Return as lowercase string if not numerical
        return answer.lower()

def compare_answers(pred, true, tolerance=1e-3):
    """
    Compare predicted and true answers with better handling
    """
    pred_norm = normalize_numerical_answer(pred)
    true_norm = normalize_numerical_answer(true)

    if pred_norm is None or true_norm is None:
        return False

    # Both are numbers
    if isinstance(pred_norm, (int, float)) and isinstance(true_norm, (int, float)):
        # Use relative tolerance for large numbers
        if abs(true_norm) > 1:
            return abs(pred_norm - true_norm) / abs(true_norm) < tolerance
        else:
            return abs(pred_norm - true_norm) < tolerance

    # String comparison (case-insensitive)
    return str(pred_norm).lower() == str(true_norm).lower()

# ============================================================
# STEP 4: Improved Model Inference Function
# ============================================================

def generate_answer(question, max_length=512, temperature=0.0):
    """
    Generate answer for a math question
    Qwen2.5-Math format - OPTIMIZED FOR SPEED
    """

    # Qwen2.5-Math uses chat template with specific format
    messages = [
        {"role": "system", "content": "Please reason step by step, and put your final answer within \\boxed{}."},
        {"role": "user", "content": question}
    ]

    # Apply chat template
    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    # Tokenize
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=2048)
    inputs = {k: v.to(model.device) for k, v in inputs.items()}

    # Generate - OPTIMIZED SETTINGS
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_length,
            temperature=temperature if temperature > 0 else None,  # Greedy decoding when temp=0
            do_sample=False,  # Greedy decoding is faster
            pad_token_id=tokenizer.pad_token_id or tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
            use_cache=True,  # Enable KV cache
        )

    # Decode only the generated part (not the input prompt)
    generated_ids = outputs[0][inputs['input_ids'].shape[1]:]
    response = tokenizer.decode(generated_ids, skip_special_tokens=True)

    return response.strip()

# ============================================================
# STEP 5: Load Dataset
# ============================================================

from datasets import load_dataset

print("Loading dataset...")
DATASET_NAME = "gsm8k"
DATASET_CONFIG = "main"
SPLIT = "test"
NUM_SAMPLES = 50  # Start small for testing

dataset = load_dataset(DATASET_NAME, DATASET_CONFIG, split=f"{SPLIT}[:{NUM_SAMPLES}]")
print(f"✅ Loaded {len(dataset)} samples from {DATASET_NAME}")

# ============================================================
# STEP 6: Run Evaluation with Better Error Handling
# ============================================================

def evaluate_model(dataset, output_dir="./eval_results", dataset_type="gsm8k", batch_size=1):
    """Run evaluation on dataset and save results"""

    os.makedirs(output_dir, exist_ok=True)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    # Enable faster inference
    model.eval()
    if torch.cuda.is_available():
        torch.backends.cudnn.benchmark = True

    results = []
    correct = 0
    total_processed = 0

    print(f"\n{'='*70}")
    print(f"Starting Evaluation on {len(dataset)} samples")
    print(f"{'='*70}\n")

    for idx, example in enumerate(tqdm(dataset, desc="Evaluating")):
        question = example['question']
        true_answer_text = example['answer']

        # Extract ground truth
        true_answer = extract_answer_from_response(true_answer_text, dataset_type)

        try:
            # Generate prediction - FASTER SETTINGS
            response = generate_answer(
                question,
                max_length=384,  # Reduced from 512 - most answers are shorter
                temperature=0.0  # Greedy decoding is faster and often better for math
            )
            predicted_answer = extract_answer_from_response(response, dataset_type)

            # Check correctness
            is_correct = compare_answers(predicted_answer, true_answer)
            if is_correct:
                correct += 1

            total_processed += 1

            result = {
                'index': idx,
                'question': question,
                'true_answer': true_answer,
                'predicted_answer': predicted_answer,
                'correct': is_correct,
                'full_response': response,
                'ground_truth_text': true_answer_text
            }

            # Print progress every 10 samples
            if (idx + 1) % 10 == 0:
                acc = (correct / total_processed * 100) if total_processed > 0 else 0
                print(f"\n--- Progress: {idx + 1}/{len(dataset)} ---")
                print(f"Current Accuracy: {acc:.2f}% ({correct}/{total_processed})")
                print(f"Last Q: {question[:60]}...")
                print(f"True: {true_answer} | Pred: {predicted_answer} | {'✅' if is_correct else '❌'}")

        except Exception as e:
            print(f"\n⚠️ Error on sample {idx}: {str(e)}")
            result = {
                'index': idx,
                'question': question,
                'true_answer': true_answer,
                'error': str(e),
                'correct': False
            }

        results.append(result)

    # Calculate final metrics
    accuracy = (correct / total_processed * 100) if total_processed > 0 else 0

    # Save results
    results_file = os.path.join(output_dir, f"results_{timestamp}.json")
    with open(results_file, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    metrics = {
        'dataset': DATASET_NAME,
        'total_samples': len(results),
        'processed': total_processed,
        'correct': correct,
        'accuracy': accuracy,
        'timestamp': timestamp
    }

    metrics_file = os.path.join(output_dir, f"metrics_{timestamp}.json")
    with open(metrics_file, 'w') as f:
        json.dump(metrics, f, indent=2)

    # Create CSV
    df = pd.DataFrame(results)
    csv_file = os.path.join(output_dir, f"results_{timestamp}.csv")
    df.to_csv(csv_file, index=False)

    # Print final results
    print("\n" + "="*70)
    print("EVALUATION COMPLETE")
    print("="*70)
    print(f"Dataset: {DATASET_NAME}")
    print(f"Total Samples: {len(results)}")
    print(f"Successfully Processed: {total_processed}")
    print(f"Correct: {correct}")
    print(f"Accuracy: {accuracy:.2f}%")
    print(f"\nResults saved to:")
    print(f"  - {results_file}")
    print(f"  - {metrics_file}")
    print(f"  - {csv_file}")
    print("="*70)

    return results, metrics

# Run evaluation
results, metrics = evaluate_model(dataset, output_dir="./eval_results", dataset_type="gsm8k")

# ============================================================
# STEP 7: Detailed Analysis
# ============================================================

print("\n" + "="*70)
print("SAMPLE PREDICTIONS (First 5)")
print("="*70)

for i, result in enumerate(results[:5]):
    if 'error' not in result:
        status = '✅ CORRECT' if result['correct'] else '❌ INCORRECT'
        print(f"\n--- Sample {i+1} {status} ---")
        print(f"Q: {result['question'][:100]}...")
        print(f"True Answer: {result['true_answer']}")
        print(f"Predicted: {result['predicted_answer']}")
        print(f"Response snippet: {result['full_response'][:150]}...")

# Analyze errors
incorrect = [r for r in results if 'correct' in r and not r['correct']]
print(f"\n\n{'='*70}")
print(f"ERROR ANALYSIS: {len(incorrect)} incorrect predictions")
print("="*70)

for i, result in enumerate(incorrect[:3]):
    print(f"\n--- Error Example {i+1} ---")
    print(f"Q: {result['question'][:100]}...")
    print(f"Expected: {result['true_answer']}")
    print(f"Got: {result['predicted_answer']}")
    print(f"Full response: {result['full_response'][:300]}...")

print("\n✅ Evaluation complete! Check the results files for full details.")
```
